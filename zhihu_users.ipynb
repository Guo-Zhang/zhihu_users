## Load Data

```python
# work with MySQLdb
db = MySQLdb.connect(host='localhost', port=3306, user='root', passwd='', db='zhihu')
cursor = db.cursor()
cursor.execute('select * from user')
data = cursor.fetchall()
df = pd.DataFrame(list(data),columns=['id','u_name','u_token','u_follows','u_thank','u_favorite','u_voteup','u_rank','u_frank'])
```

```python
# work with sqlalchemy
url = URL('mysql',host='localhost', port=3306, username='root', password='',database='zhihu')
engine = create_engine(url)  # dialect+driver://username:password@host:port/database
connection = engine.connect()
data = connection.execute('select * from user')
df = pd.DataFrame(list(data),columns=['id','u_name','u_token','u_follows','u_thank','u_favorite','u_voteup','u_rank','u_frank'])
```

### Summary

```python
# all users
df[['u_follows','u_thank','u_favorite','u_voteup']].describe()
```

```python
# user with no less then 1000 
df2 = df[df['u_follows']>=1000]
df2[['u_follows','u_thank','u_favorite','u_voteup']].describe()
```

```python
# users with high followers number but low thank, favorite and voteup
df_outlier = df[(df['u_follows']>1000) & (df['u_voteup']<1)&(df['u_favorite']<1)&(df['u_thank']<1)]
len(df_outlier)
```

```python
df_outlier.sort_values('u_follows',ascending=False)
```

### Regression

```python
import statsmodels.api as sm
```

```python
# all users
x = sm.add_constant(df[['u_thank','u_favorite','u_voteup']])
y = df['u_follows']
model = sm.OLS(y,x)
results = model.fit()
results.summary()
```

```python
# users with no less than 1000 followers
x = sm.add_constant(df2[['u_thank','u_favorite','u_voteup']])
y = df2['u_follows']
model = sm.OLS(y,x)
results = model.fit()
results.summary()
```
